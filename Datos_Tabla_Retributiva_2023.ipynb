```python
#Creamos la columna nueva calculando la media de las filas de forma independiente
#Utilizamos una lista de tuplas con el inicio y el fin de cada posición de la fila, de esta forma calculamos la media independiente con un bucle for
#Para crear la media independiente, creamos dentro del bucle una variable media a la que le pasamos el inicio y el fin que iteramos con el for y calculamos su media
#Asignamos los valores de la variable creada anteriormente a la columna nueva creada en el df

rangos_filas = [(0, 10), (56, 59), (59, 98), (98, 101), (101, 119), (119, 136), (136, 138),
                (138, 144), (144, 187), (187, 206), (206, 242), (242, 278), (278, 292),
                (292, 300), (300, 356), (356, 383), (383, 386), (386, 436), (436, 456),
                (456, 523), (523, 532), (532, 585), (585, 626)]

df["media_compl_especif"] = None


for inicio, fin in rangos_filas:
    media = df.iloc[inicio:fin, 3].mean()
    df.loc[inicio:fin-1, "media_compl_especif"] = media
```
